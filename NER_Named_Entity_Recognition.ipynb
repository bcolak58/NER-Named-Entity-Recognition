```python
from spacy import displacy
```

```python
NER=spacy.load('en_core_web_sm')
```

```python
text="""Bahri COLAK live in Turkey and he loves Artifical Intelligence. Python is a great programming language.
Turkis Space Agency is founded in 2019 in Ankara"""
```

```python
text1=NER(text)
```

```python
for word in text1.ents:
    print(word.text,word.label_)
```
